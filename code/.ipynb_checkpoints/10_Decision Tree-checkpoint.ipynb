```python
# Import Libraries
import pandas as pd
from sklearn import tree
import graphviz as gv
```

```python
# load and Explore Data
car = pd.read_csv('C:/Users/felix.schildorfer/OneDrive - Avanade/Education/ML/Statistical-Learning-in-Python/data/Carseats.csv')
car.head()
```

```python
car.describe()
```

```python
# Create a list to store the data
high = []

# For each row in the column,
for row in car['Sales']:
    # if more than a value,
    if row <= 8:
        high.append(1)
    # else, if more than a value,
    else:
        high.append(0)
        
# Create a column from the list
car['High'] = high

# Change data to eliminate Strings
car['Urban'] = car['Urban'].map({'Yes': 1, 'No': 0})
car['US'] = car['US'].map({'Yes': 1, 'No': 0})
car['ShelveLoc'] = car['ShelveLoc'].map({'Medium': 1, 'Bad': 0, 'Good': 3})
```

```python
# Deploy Decison Tree
X = car.iloc[:, 1:11]
y = car.iloc[:, 12]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
# Graph Tree
dot_data = tree.export_graphviz(clf, out_file=None,  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = gv.Source(dot_data)
#graph
```

```python
# Predict result
clf.predict_proba(car.iloc[:, 12])
```
